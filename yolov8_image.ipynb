2. 필요한 라이브러리 임포트

```python
from ultralytics import YOLO
from PIL import Image
import requests
import io
```

3. 사전 훈련된 YOLOv8 모델 로드


```python
model = YOLO('yolov8x.pt')
```

4. 인터넷에서 도로 이미지 가져오기

```python
url = "https://www.gimhae.go.kr/CmsMultiFile/view.do?multifileId=NS52310&idx=18883&q=100" # 원하는 도로 이미지 URL로 변경하세요
response = requests.get(url)
img = Image.open(io.BytesIO(response.content))
```

5. 객체 감지 실행

```python
results = model(img)
```

6. 결과 시각화

```python
results[0].plot()
```
